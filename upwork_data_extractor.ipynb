```python
from parsel import Selector
import pandas as pd
```

```python
file_list = []
for i in range(1,81):
    file_list.append(f"{i}.txt")
```

```python
# List of columns
title = []
price_list = []
expertise_list = []
number_of_question_list = []
questions_list = []

country_list = []
member_since_list = []
job_posted_list = []
payment_list = []
review_list = []
total_spent_list = []
jobs_prgress_list = []
```

```python
for i in file_list:
    with open(i, 'r', encoding="utf8") as file:
        html = file.read().replace('\n', '')
        response=Selector(text=html)
        job_title = response.xpath('//h1[@class="m-0 h2"]/text()').get()
        price = response.xpath('///div[@data-cy="fixed-price"]/../strong/text()').get()
        expertise = response.xpath('///div[@data-cy="expertise"]/../strong/text()').get()
        questions = response.xpath('//p[@class="mb-5"]/../ol/li/text()').getall()
        question_number = response.xpath('//p[@class="mb-5"]/../ol/li')
        if question_number is not None:
            number_of_question_list.append(len(question_number))
        
        
        
        country = response.xpath('//li[@data-qa="client-location"]/strong/text()').get()
        member_since = response.xpath('//li[@data-qa="client-contract-date"]/small/text()').get()
        job_posted = response.xpath('normalize-space(//li[@data-qa="client-job-posting-stats"]/strong/text())').get()
        payment = response.xpath('//div[@class="payment-verified mr-5 up-icon up-popper-trigger"]/../strong/text()').get()
        review = response.xpath('//div[@class="up-rating vertical-align-text-bottom"]/../span[@class="nowrap"]/text()').get()
        total_spent = response.xpath('//strong[@data-qa="client-spend"]/span/text()').get()
        jobs_prgress = response.xpath('//a[@class="js-job-link"]/text()').getall()
        

        title.append(job_title)
        price_list.append(price)
        expertise_list.append(expertise)
        questions_list.append(questions)
        
        country_list.append(country)
        member_since_list.append(member_since)
        job_posted_list.append(job_posted)
        payment_list.append(payment)
        review_list.append(review)
        total_spent_list.append(total_spent)
        jobs_prgress_list.append(jobs_prgress)
        
        
        
```

```python
data = {
    "title": title,
    "price": price_list,
    "expertise": expertise_list,
    "questions": questions_list,
    "q_number": number_of_question_list,
    "country": country_list,
    "member_since": member_since_list,
    "job_posted": job_posted_list,
    "payment": payment_list,
    "review": review_list,
    "spent": total_spent_list,
    "jobs_progress": jobs_prgress_list
    
}




df = pd.DataFrame(data)
df.to_csv("data.csv", index=False)
```
