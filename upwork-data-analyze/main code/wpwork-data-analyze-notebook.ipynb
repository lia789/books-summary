```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("upwork_data.csv")
df.shape
```

```python
missing_jobs = df[df['upwork_category'].isnull()]
missing_jobs_urls= list(missing_jobs["URLs"])
missing_jobs.shape
```

```python
# Missing client location urls

missing_client_location = df[df['client_location'].isnull()]
client_location_list = list(missing_client_location["URLs"])
len(client_location_list)
```

```python
# Droping Unwanted data
df.dropna(subset=['upwork_category'],inplace=True)

# Dataset shape
df.shape
```
